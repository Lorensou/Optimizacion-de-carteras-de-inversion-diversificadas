## Genero 100K carteras aleatorias

```python
import random


def generar_carteras_aleatorias(titulos, n_carteras):
    rendimientos = [random.uniform(0, 0.36) for _ in range(len(titulos))]
    volatilidades = [random.uniform(0, 0.60) for _ in range(len(titulos))]
    carteras_aleatorias = []
    
    for _ in range(n_carteras):
        pesos = [random.uniform(0, 1) for _ in range(len(titulos))]
        suma_pesos = sum(pesos)
        pesos_normalizados = [peso / suma_pesos for peso in pesos]
        
        # Calcular el rendimiento y la volatilidad de la cartera actual
        rendimiento = np.dot(pesos_normalizados, rendimientos)
        volatilidad = np.sqrt(np.dot(pesos_normalizados, np.multiply(volatilidades, volatilidades)))
        
        carteras_aleatorias.append({'Rendimiento': rendimiento , 'Volatilidad': volatilidad})
    
    df_carteras = pd.DataFrame(carteras_aleatorias)
    
    return df_carteras


```

```python
titulos = ['Ibex', 'S&P', 'DJ', 'NASDAQ_COMP', 'NASDAQ_100', 'FTSE', 'CAC', 'EUSTOXX', 'GDA', 'NIKKEI', 'NIFTY', 'HANGSENG']
n_carteras = 100000

carteras_aleatorias = generar_carteras_aleatorias(titulos, n_carteras)

carteras_aleatorias
```
